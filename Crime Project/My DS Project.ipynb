#### Import Packages :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
from pandasql import sqldf
```

#### Read CSV File :

```python
df_22 = pd.read_csv('2022.csv')
df_22.index.name = 'Index'
df_22.rename(columns = {'OFFENSE_DESCRIPTION':'Offense', 'YEAR':'Year', 'OFFENSE_CODE':'Code'}, inplace = True)

df_21 = pd.read_csv('2021.csv', dtype = object)
df_21.index.name = 'Index'
df_21.rename(columns = {'OFFENSE_DESCRIPTION':'Offense', 'YEAR':'Year', 'OFFENSE_CODE':'Code'}, inplace = True)

df_20 = pd.read_csv('2020.csv', dtype = object)
df_20.index.name = 'Index'
df_20.rename(columns = {'OFFENSE_DESCRIPTION':'Offense', 'YEAR':'Year', 'OFFENSE_CODE':'Code'}, inplace = True)

df_19 = pd.read_csv('2019.csv', dtype = object)
df_19.index.name = 'Index'
df_19.rename(columns = {'OFFENSE_DESCRIPTION':'Offense', 'YEAR':'Year', 'OFFENSE_CODE':'Code'}, inplace = True)

mysql = lambda q: sqldf(q, globals())
```

#### Get specific crimes using SQL :

```python
query = '''
SELECT Offense, Year, Lat, Long FROM df_19
WHERE Offense LIKE 'LARCENY%' AND Lat <> 0
UNION ALL
SELECT Offense, Year, Lat, Long FROM df_20
WHERE Offense LIKE 'LARCENY%' AND Lat <> 0
UNION ALL
SELECT Offense, Year, Lat, Long FROM df_21
WHERE Offense LIKE 'LARCENY%' AND Lat <> 0
UNION ALL
SELECT Offense, Year, Lat, Long FROM df_22
WHERE Offense LIKE 'LARCENY%' AND Lat <> 0
ORDER BY Offense DESC;
'''
larceny = mysql(query)
larceny
```
