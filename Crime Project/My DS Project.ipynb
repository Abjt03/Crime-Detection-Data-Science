#### Read CSV File :

#### Clean the CSV Files :

```python
query = '''
SELECT Code, Offense, Year, Month, Lat, Long
FROM df_19
WHERE Lat <> 0 AND Code NOT IN (612
UNION ALL
SELECT Code, Offense, Year, Month, Lat, Long
FROM df_20
WHERE Lat <> 0 AND Code <> 612
UNION ALL
SELECT Code, Offense, Year, Month, Lat, Long
FROM df_21
WHERE Lat <> 0 AND Code <> 612
UNION ALL
SELECT Code, Offense, Year, Month, Lat, Long
FROM df_22
WHERE Lat <> 0 AND Code <> 612
ORDER BY Year DESC
'''
Clean_Boston_Crime_Data = mysql(query)
Clean_Boston_Crime_Data
```

```python
Clean_Boston_Crime_Data = Clean_Boston_Crime_Data.astype({'Code':'int32', 'Offense':'category', 'Year':'int32', 'Month':'int32', 'Lat':'float64', 'Long':'float64'})
Clean_Boston_Crime_Data.Offense = Clean_Boston_Crime_Data.Offense.str.capitalize()
Clean_Boston_Crime_Data.info()
```
