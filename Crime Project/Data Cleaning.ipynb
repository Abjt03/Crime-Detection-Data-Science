#### Read CSV File :

```python
df_22 = pd.read_csv('2022.csv')
df_22.index.name = 'Index'
df_22.columns = df_22.columns.str.capitalize()

df_21 = pd.read_csv('2021.csv', dtype = object)
df_21.index.name = 'Index'
df_21.columns = df_22.columns.str.capitalize()

df_20 = pd.read_csv('2020.csv', dtype = object)
df_20.index.name = 'Index'
df_20.columns = df_22.columns.str.capitalize()

df_19 = pd.read_csv('2019.csv', dtype = object)
df_19.index.name = 'Index'
df_19.columns = df_22.columns.str.capitalize()

mysql = lambda q: sqldf(q, globals())
```

#### Clean the CSV Files :

```python
query = '''
SELECT Code, Offense, Date, Street, Year, Month, Lat, Long
FROM df_19
WHERE Lat <> 0 AND Code NOT IN (612, 99999, 1501, 3125)
UNION ALL
SELECT Code, Offense, Date, Street, Year, Month, Lat, Long
FROM df_20
WHERE Lat <> 0 AND Code NOT IN (612, 99999, 1501, 3125)
UNION ALL
SELECT Code, Offense, Date, Street, Year, Month, Lat, Long
FROM df_21
WHERE Lat <> 0 AND Code NOT IN (612, 99999, 1501, 3125)
UNION ALL
SELECT Code, Offense, Date, Street, Year, Month, Lat, Long
FROM df_22
WHERE Lat <> 0 AND Code NOT IN (612, 99999, 1501, 3125)
ORDER BY Year DESC
'''
Clean_Boston_Crime_Data = mysql(query)
Clean_Boston_Crime_Data.Date = pd.to_datetime(Clean_Boston_Crime_Data.Date).dt.normalize()
Clean_Boston_Crime_Data
```

```python
Clean_Boston_Crime_Data = Clean_Boston_Crime_Data.astype({'Code':'int32', 'Offense':'string', 'Year':'int32', 'Month':'int32', 'Lat':'float64', 'Long':'float64'})
Clean_Boston_Crime_Data.Offense = Clean_Boston_Crime_Data.Offense.str.capitalize()
Clean_Boston_Crime_Data.info()
```

#### Ensuring each crime is mapped to a unique code:

```python
Clean_Boston_Crime_Data.Offense = Clean_Boston_Crime_Data.Offense.replace(['Murder, non-negligent manslaughter', 'Robbery - street', 'Burglary - residential', 'Burglary - commerical', 'Migrated report - stolen property', 'Drugs - sale / manufacturing', 'Val - violation of auto law - other', 'Property - lost'], ['Murder, non-negligient manslaughter', 'Robbery', 'Burglary - residential - force', 'Burglary - commerical - force', 'Stolen property - buying / receiving / possessing', 'Drugs - possession/ sale/ manufacturing/ use', 'Migrated report - auto law violation', 'Property - lost/ missing'])
```

```python
query = '''
SELECT Code, Offense FROM Clean_Boston_Crime_Data
GROUP BY Offense
ORDER BY Code
'''
mysql(query)
```

#### Convert final DataFrame to CSV

```python
Clean_Boston_Crime_Data.to_csv('Clean_Boston_Crime_Data.csv', index = False)
```
